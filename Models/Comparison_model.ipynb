```python
import pandas as pd

results = [
    ["Random Forest", 1.00, 0.91, 0.95, 0.99],
    ["XGBoost", 1.00, 0.91, 0.95, 0.99],
    ["LightGBM", 1.00, 0.91, 0.93, 0.99],
    ["SVM", 0.99, 0.91, 0.90, 0.97],
    ["Isolation Forest", 0.98, 0.61, 0.62, 0.90],
    ["Auto Encoders",0.95,0.89,0.51,0.89],
    ["LSTM",0.91339,0.94,0.92,0.98],
    ["Transformer", 0.9965, 0.9383, 0.9560, 0.9923]
]

df_results = pd.DataFrame(
    results,
    columns=["Model","Accuracy","Recall","F1","AUC"]
)

print(df_results)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

fig, axes = plt.subplots(2,2, figsize=(14,10))

sns.barplot(x="Model", y="Accuracy", data=df_results, ax=axes[0,0])
axes[0,0].set_title("Model Accuracy")
axes[0,0].tick_params(axis='x', rotation=45)

sns.barplot(x="Model", y="Recall", data=df_results, ax=axes[0,1])
axes[0,1].set_title("Model Recall")
axes[0,1].tick_params(axis='x', rotation=45)

sns.barplot(x="Model", y="F1", data=df_results, ax=axes[1,0])
axes[1,0].set_title("Model F1 Score")
axes[1,0].tick_params(axis='x', rotation=45)

sns.barplot(x="Model", y="AUC", data=df_results, ax=axes[1,1])
axes[1,1].set_title("Model ROC-AUC")
axes[1,1].tick_params(axis='x', rotation=45)

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Style
sns.set_style("whitegrid")
sns.set_context("talk")

fig, axes = plt.subplots(2,2, figsize=(16,12))

# Accuracy
sns.barplot(
    x="Model", y="Accuracy",
    data=df_results,
    palette="Blues_d",
    ax=axes[0,0]
)
axes[0,0].set_title("Accuracy Comparison", fontsize=16)
axes[0,0].tick_params(axis='x', rotation=35)

# Recall
sns.barplot(
    x="Model", y="Recall",
    data=df_results,
    palette="Greens_d",
    ax=axes[0,1]
)
axes[0,1].set_title("Recall Comparison", fontsize=16)
axes[0,1].tick_params(axis='x', rotation=35)

# F1 Score
sns.barplot(
    x="Model", y="F1",
    data=df_results,
    palette="Purples_d",
    ax=axes[1,0]
)
axes[1,0].set_title("F1 Score Comparison", fontsize=16)
axes[1,0].tick_params(axis='x', rotation=35)

# AUC
sns.barplot(
    x="Model", y="AUC",
    data=df_results,
    palette="Oranges_d",
    ax=axes[1,1]
)
axes[1,1].set_title("ROC-AUC Comparison", fontsize=16)
axes[1,1].tick_params(axis='x', rotation=35)

plt.tight_layout()
for ax in axes.flatten():
    for container in ax.containers:
        ax.bar_label(container, fmt="%.2f")
plt.show()
```

```python
print("Model Comparison & Selection Rationale:")
print("We evaluated multiple Machine Learning and Deep Learning models on anomaly detection using key performance metrics: Accuracy, Recall, F1-score, and AUC. ")
print("Random Forest, XGBoost, and LightGBM achieved near-perfect accuracy (~1.00) and very high AUC (~0.99), indicating strong classification capability. However, such extremely high accuracy may suggest overfitting, especially in anomaly detection tasks where class imbalance exists. ")
print("SVM performed slightly lower but remained stable, though its F1-score indicates it is less effective in balancing precision and recall.")
print("Isolation Forest and Autoencoders showed comparatively lower recall and F1-score, meaning they are less reliable in detecting anomalies consistently.")
print("Deep Learning models (LSTM and Transformer) demonstrated the most balanced and reliable performance:")
print("LSTM: High recall (0.94) and strong F1-score (0.92), making it effective at capturing sequential anomaly patterns.")
print("Transformer: Best overall balance with high accuracy (~0.996), recall (~0.938), F1 (~0.956), and AUC (~0.992), indicating strong generalization.")
```
